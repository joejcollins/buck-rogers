To find the biggest of two numbers
you simply have to compare them with each other.
This is direct and obvious approach that is easy to understand.

## Max Logic

This can be packaged as a function to aid reuse.

```python
"""Functions to find the maximum of two numbers."""
import ipytest

ipytest.autoconfig()

def find_max_logic(a, b):
    """Find the maximum of two numbers using logic."""
    if a > b:
        return a
    return b
```

## Tests

It is important to test that the function works.

But because I am lazy I will not test any of the boundary conditions,
like if both numbers are the same
or if one number is replaced with a letter.

```python
def test_find_max_logic():
    """Test the find_max_logic function."""
    # Arrange
    five = 5
    ten = 10
    # Act & Assert
    assert find_max_logic(3, five) == five
    assert find_max_logic(ten, 2) == ten
```
